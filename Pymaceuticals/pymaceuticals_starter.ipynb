## Observations and Insights

Of the four high-performing regminens, it appears that Ramicane was the best performer. It had the lowest median tumor volume of 36.56mm3, while its next-nearest competitor, Capomulin, had a median tumor volume of 38.12mm3. Infubinol and Ceftamin performed particularly poorly with a median tumor volume of 60.17mm3, or about 1.6 times the median-size of tumor in a mouse treated with Ramicane.

The efficacy of Capomulin on Mouse i557 seems to be largely limited to the first 5 to 7 days of treatment. There was a sharp decline in tumor volume during that period, after which the tumor seems to start a period of high growth.

There is a strong, positive correlation between the weight of a mouse and the size of its tumor, as evidenced by the regression analysis displayed in the last figure.

## Quartiles, outliers and boxplots

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens. 

# Group the data on Mouse ID, Drug Regimen, and Tumor Vol
treatment_df = combined_df.loc[(combined_df["Drug Regimen"] == "Capomulin") 
                               | (combined_df["Drug Regimen"] == "Ramicane")
                               | (combined_df["Drug Regimen"] == "Infubinol") 
                               | (combined_df["Drug Regimen"] == "Ceftamin") 
                               ,["Mouse ID","Drug Regimen","Tumor Volume (mm3)"]]

grouped_treatment_df = treatment_df.groupby(["Drug Regimen", "Mouse ID"]).last()
sorted_treatment_df = grouped_treatment_df.sort_values(by = "Tumor Volume (mm3)").reset_index()
renamed_df = sorted_treatment_df.rename(columns = {"Tumor Volume (mm3)":"Final Tumor Volume"})
renamed_df
```

```python
# Calculate the IQR and quantitatively determine if there are any potential outliers. 


# Insert final tumor volume to a var
final_tumor_vol_capomulin = renamed_df.loc[renamed_df["Drug Regimen"] == "Capomulin", "Final Tumor Volume"]
final_tumor_vol_ramicane = renamed_df.loc[renamed_df["Drug Regimen"] == "Ramicane","Final Tumor Volume"]
final_tumor_vol_infubinol = renamed_df.loc[renamed_df["Drug Regimen"] == "Infubinol","Final Tumor Volume"]
final_tumor_vol_ceftamin = renamed_df.loc[renamed_df["Drug Regimen"] == "Ceftamin","Final Tumor Volume"]


############################################
# Run statistics for Capomulin
############################################
# Calculate IQR Capomulin
quartiles_capomulin = final_tumor_vol_capomulin.quantile([.25, .5, .75])
lowerq_capomulin = quartiles_capomulin[0.25]
upperq_capomulin = quartiles_capomulin[0.75]
iqr_capomulin = upperq_capomulin - lowerq_capomulin


# Calculate upper and lower bounds
upper_bound_capomulin = upperq_capomulin + (1.5 * iqr_capomulin) # 94.04403868375002 data set has hi of 47.68596303
lower_bound_capomulin = lowerq_capomulin - (1.5 * iqr_capomulin) # 3.0739668737499812 data set has low of 23.34359787


# final_tumor_vol_capomulin
if final_tumor_vol_capomulin.max() > upper_bound_capomulin:
    print("There is an outlier above the upper bound in the Capomulin data set.")
elif final_tumor_vol_capomulin.min() < lower_bound_capomulin:
    print("There is an outlier below the lower bound in the Capomulin data set.")
elif final_tumor_vol_capomulin.max() > upper_bound_capomulin and final_tumor_vol_capomulin.min() < lower_bound_capomulin:
    print("There are outliers on the high and low side of the Capomulin set.")
else: 
    print("There are no outliers in the Capomulin data set!")
    
# ############################################
# # Run statistics for Ramicane
# ############################################
# # Calculate IQR Ramicane
quartiles_ramicane = final_tumor_vol_ramicane.quantile([.25, .5, .75])
lowerq_ramicane = quartiles_ramicane[0.25]
upperq_ramicane = quartiles_ramicane[0.75]
iqr_ramicane = upperq_ramicane - lowerq_ramicane

# Calculate upper and lower bounds
upper_bound_ramicane = upperq_ramicane + (1.5 * iqr_ramicane) #94.04403868375002
lower_bound_ramicane = lowerq_ramicane - (1.5 * iqr_ramicane) #3.0739668737499812


# final_tumor_vol_capomulin
if final_tumor_vol_ramicane.max() > upper_bound_ramicane:
    print("There is an outlier above the upper bound in the ramicane data set.")
elif final_tumor_vol_ramicane.min() < lower_bound_ramicane:
    print("There is an outlier below the lower bound in the ramicane data set.")
elif final_tumor_vol_ramicane.max() > upper_bound_ramicane and final_tumor_vol_ramicane.min() < lower_bound_ramicane:
    print("There are outliers on the high and low side of the Ramicane set.")
else: 
    print("There are no outliers in the ramicane data set!")
    

############################################
# Run statistics for Infubinol
############################################
# Calculate IQR infubinol
quartiles_infubinol = final_tumor_vol_infubinol.quantile([.25, .5, .75])
lowerq_infubinol = quartiles_infubinol[0.25]
upperq_infubinol = quartiles_infubinol[0.75]
iqr_infubinol = upperq_infubinol - lowerq_infubinol

# Calculate upper and lower bounds
upper_bound_infubinol = upperq_infubinol + (1.5 * iqr_infubinol) #
lower_bound_infubinol = lowerq_infubinol - (1.5 * iqr_infubinol) #


if final_tumor_vol_infubinol.max() > upper_bound_infubinol:
    print("There is an outlier above the upper bound in the infubinol data set.")
elif final_tumor_vol_infubinol.min() < lower_bound_infubinol:
    print("There is an outlier below the lower bound in the infubinol data set.")
elif final_tumor_vol_infubinol.max() > upper_bound_infubinol and final_tumor_vol_infubinol.min() < lower_bound_infubinol:
    print("There are outliers on the high and low side of the Infubinol set.")
else: 
    print("There are no outliers in the infubinol data set!")
    
    
############################################
# Run statistics for Ceftamin
############################################
# Calculate IQR ceftamin
quartiles_ceftamin = final_tumor_vol_ceftamin.quantile([.25, .5, .75])
lowerq_ceftamin = quartiles_ceftamin[0.25]
upperq_ceftamin = quartiles_ceftamin[0.75]
iqr_ceftamin = upperq_ceftamin - lowerq_ceftamin

# Calculate upper and lower bounds
upper_bound_ceftamin = upperq_ceftamin + (1.5 * iqr_ceftamin) #
lower_bound_ceftamin = lowerq_ceftamin - (1.5 * iqr_ceftamin) #


if final_tumor_vol_ceftamin.max() > upper_bound_ceftamin:
    print("There is an outlier above the upper bound in the ceftamin data set.")
elif final_tumor_vol_ceftamin.min() < lower_bound_ceftamin:
    print("There is an outlier below the lower bound in the ceftamin data set.")
elif final_tumor_vol_ceftamin.max() > upper_bound_ceftamin and final_tumor_vol_ceftamin.min() < lower_bound_ceftamin:
    print("There are outliers on the high and low side of the Ceftamin set.")
else: 
    print("There are no outliers in the ceftamin data set!")
    
# print(f"cap:{iqr_capomulin}, ram:{iqr_ramicane}, inf:{iqr_infubinol}, cef:{iqr_ceftamin}")
# print(f"cap:{quartiles_capomulin} ram: {quartiles_ramicane} inf{quartiles_infubinol} ceft: {quartiles_ceftamin}")
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen

# Group mouse data by Mouse ID and weight
capo_mouse_grouped = capo_mouse.groupby(["Mouse ID", "Weight (g)"])

# Calculate the mean for the group mice
capo_mouse_avg_vol= capo_mouse_grouped["Tumor Volume (mm3)"].mean()

# Create DF with new Mean data
capo_scatter_df = pd.DataFrame({"Avg_Tumor_Vol": capo_mouse_avg_vol
                               })
capo_scatter_df.reset_index(level=['Weight (g)'], inplace=True)

# Build the scatter plot
x_values = capo_scatter_df["Weight (g)"]
y_values = capo_scatter_df["Avg_Tumor_Vol"]

plt.scatter(x_values, y_values)
plt.xlabel("Mouse Weight")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```

```python
# Calculate the correlation coefficient and linear regression model for mouse weight and average tumor 
# volume for the Capomulin regimen

# Set x and y values
x_values = capo_scatter_df["Weight (g)"]
y_values = capo_scatter_df["Avg_Tumor_Vol"]

# Run regression
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# Create Scatter plot
plt.scatter(x_values, y_values)
plt.plot(x_values,regress_values,"r-")

# Style the plot
plt.annotate(line_eq,(20,36) ,fontsize = 12, color = "red")
plt.xlabel("Mouse Weight")
plt.ylabel("Average Tumor Size (mm3)")
plt.show()
```
